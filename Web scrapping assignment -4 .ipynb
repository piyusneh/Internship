#### Q 1) Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
#import required libraries
from selenium import webdriver
import time
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
```

```python
#connecting to the chrome driver
driver=webdriver.Chrome(r"C://Chromedriver.exe")
driver.maximize_window()
```

```python
# pass the url of wikipedia
url= "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
#scroll down the page
driver.execute_script("window.scrollTo(0, 1000)")  #1000 is the height that needs to be scrolled
```

```python
#finding details
details=[]
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td"):
    details.append(i.text.replace("\n"," "))
details[0:6]
```

```python
#seperating the details
#creating empty list
rank=[]
name=[]
artist=[]
views=[]
upload=[]
count=0
#seperating details using indexing
while count<int(len(details)):
    rank.append(details[count])
    name.append(details[count+1])
    artist.append(details[count+2])
    views.append(details[count+3])
    upload.append(details[count+4])
    count=count+6 
print(rank[0:4])
print(name[0:4])
print(artist[0:4])
print(views[0:4])
print(upload[0:4])
```

```python
#check the length of each
print(len(rank))
print(len(name))
print(len(artist))
print(len(views))
print(len(upload))
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Rank"]=rank
df["Name"]=name
df["Artist"]=artist
df['Upload date']=upload
df['Views']=views
df
```

```python
#create empty list
match_title=[]
place=[]
series=[]
date=[]
time=[]
#match title
for title in driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p"):
    match_title.append(title.text.split("\n")[0])
#place    
for s in driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p"):
    place.append(s.text.split("\n")[1])
#series
for s in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[2]"):
    series.append(s.text)        
#date
for d in driver.find_elements_by_xpath("//div[@class='fixture__full-date']"):
    date.append(d.text.split("\n")[0:2])
#time
for d in driver.find_elements_by_xpath("//div[@class='fixture__full-date']"):
    time.append(d.text.split("\n")[-1])

print(match_title)
print(place)
print(series)
print(date)
print(time)
```

```python
#check length
print(len(match_title))
print(len(series))
print(len(place))
print(len(date))
print(len(time))
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Match Title"]=match_title
df["Series"]=series
df["Place"]=place
df["Date"]=date
df["Time"]=time
df
```

#### Q 5) Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.


```python
#pass the webpage url
driver.get("https://github.com/")
```

```python
#from explore go to trending option
trending=driver.find_element_by_xpath("//details[@class='HeaderMenu-details details-overlay details-reset width-full']/div/ul[2]/li[3]/a").get_attribute("href") 
print(trending)
driver.get(trending)
```

```python
#creating empty list to store scrapped data
repository_title=[]
repository_description=[]
contributors_count=[]
language_used=[]
```

```python
#respository title
for title in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a"):
    repository_title.append(title.text)
print(repository_title[0:3])
```

```python
#for further deatils we will visit each repositorys url seperately
#fetching url of each repository
url_list=[]
for url in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a"):
    url_list.append(url.get_attribute("href"))
url_list[0:3]
```

```python
#visit each repository
for i in url_list:
    driver.get(i)
    #repository description
    try:
        description=driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/p")
        repository_description.append(description.text)
    except NoSuchElementException as e:
        repository_description.append("---")
     #contributors count  #use contains in xpath to fetch only contributors
    try:
        count=driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[5]/div/h2/a[contains(text(),'Contributors ')]")
        contributors_count.append(count.text.split(" ")[1])
    except NoSuchElementException as e:
        contributors_count.append("---")
    #language used
    try:
        lang=driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[6]/div/ul")
        language_used.append(lang.text.replace("\n"," "))
    except NoSuchElementException as e:
        language_used.append("---")
        
print(repository_description[0:2])
print(contributors_count[0:2])
print(language_used[0:2])
```

```python
#check length of each
print(len(repository_title))
print(len(repository_description))
print(len(contributors_count))
print(len(language_used))
```

```python
#create dataframe
df=pd.DataFrame({})
df["Repository title"]=repository_title
df["Repository description"]=repository_description
df["Contributors count"]=contributors_count
df["Language used"]=language_used
df
```

#### Q 10) Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/

You have to find the following details:

A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

```python
#pass the webpage url
driver.get("https://archive.ics.uci.edu/")
```

```python
#click on view all datasets
view_all=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a").get_attribute("href")
print(view_all)
driver.get(view_all)
```

```python
#create empty lists
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
number_of_attributes=[]
number_of_instances=[]
year=[]

#dataset name
for names in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a"):
    dataset_name.append(names.text)
#data type
for dtype in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p"):
    data_type.append(dtype.text)
#task
for t in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]"):
    task.append(t.text)
#type of attribute    
for attribute in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p"):
    attribute_type.append(attribute.text)
#number of attributes
for attribute_count in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]"):
    number_of_attributes.append(attribute_count.text)
#number_of_instances
for instances in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]"):
    number_of_instances.append(instances.text)
#year
for y in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]"):
    year.append(y.text)

print(dataset_name[0:2])
print(data_type[0:2])
print(task[0:2])
print(attribute_type[0:2])
print(number_of_attributes[0:2])
print(number_of_instances[0:2])
print(year[0:2])
```

```python
#checking the lengths
print(len(dataset_name))
print(len(data_type))
print(len(task))
print(len(attribute_type))
print(len(number_of_attributes))
print(len(number_of_instances))
print(len(year))
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Dataset Name"]=dataset_name
df["Data Type"]=data_type[1:] #slicing list at 1 as it includes the heading of column
df["Task"]=task[1:]
df["Attribute type"]=attribute_type[1:]
df["No of instances"]=number_of_instances[1:]
df["No of attribute"]=number_of_attributes[1:]
df["Year"]=year[1:]
df
```
